- Missing values detected in Age, Cabin, Embarked  



```python
# Fill missing age values with median
df['Age'].isna().sum()
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Drop column with excessive missing data
df = df.drop(columns=['Cabin'])
```
